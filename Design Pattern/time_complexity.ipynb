```python
import numpy as np
import matplotlib.pyplot as plt

# Define input size (n)
n = np.linspace(1, 10, 100)

# Define complexity functions
O_1 = np.ones_like(n)  # O(1) - Constant
O_log_n = np.log(n)  # O(log n) - Logarithmic
O_n = n  # O(n) - Linear
O_n_log_n = n * np.log(n)  # O(n log n) - Linearithmic
O_n2 = n**2  # O(n^2) - Quadratic
O_2n = 2**n  # O(2^n) - Exponential

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(n, O_1, label="O(1) - Constant", linestyle="--")
plt.plot(n, O_log_n, label="O(log n) - Logarithmic")
plt.plot(n, O_n, label="O(n) - Linear")
plt.plot(n, O_n_log_n, label="O(n log n) - Linearithmic")
plt.plot(n, O_n2, label="O(n²) - Quadratic")
plt.plot(n, O_2n, label="O(2ⁿ) - Exponential", linestyle="dotted", color="red")

# Labels and legend
plt.xlabel("Input Size (n)")
plt.ylabel("Operations (Time Complexity)")
plt.title("Growth of Different Time Complexities")
plt.ylim(0, 100)  # Limit y-axis for better visibility
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
